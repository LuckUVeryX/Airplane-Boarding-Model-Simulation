```python
import os
import pandas as pd

folders = ["bag_5_seat_2/","bag_5_seat_4/","bag_5_seat_0/","bag_10_seat_2/","bag_0_seat_2/"]
folder_list = []
bag_default_seat_default = os.listdir(folders[0])
folder_list.append(bag_default_seat_default)
bag_default_seat_max = os.listdir(folders[1])
folder_list.append(bag_default_seat_max)
bag_default_seat_min = os.listdir(folders[2])
folder_list.append(bag_default_seat_min)
seat_default_bag_max = os.listdir(folders[3])
folder_list.append(seat_default_bag_max)
seat_default_bag_min = os.listdir(folders[4])
folder_list.append(seat_default_bag_min)

file_list = []
for folder in folder_list:
    inner = []
    for file in folder:
        inner.append(file)
    file_list.append(inner)
file_list
```

```python
excel_list = []
for filegroup in range(len(file_list)):
    inner_2 = []
    for file in file_list[filegroup]:
        df = pd.read_csv(str(folders[filegroup])+file)
        inner_2.append(df)
    excel_list.append(inner_2)
#excel_list[0]
```

```python
excel_merge = []
for df_groups in excel_list:
    result = pd.concat(df_groups, axis=1, sort=False)
    excel_merge.append(result)
excel_merge[0]
```

```python
excel_reorder = []
for df_merge in excel_merge:
    excel_reorder.append(pd.concat([df_merge["totalBoardingTime"],df_merge["avgBoardingTime"],df_merge["seatShuffle"],df_merge["numStops"],df_merge["cumWaitTime"]],axis=1, sort=False))
```

```python
stats_list = ["totalBoardingTime","totalBoardingTime","totalBoardingTime","totalBoardingTime","totalBoardingTime","totalBoardingTime","totalBoardingTime","totalBoardingTime",
              "avgBoardingTime","avgBoardingTime","avgBoardingTime","avgBoardingTime","avgBoardingTime","avgBoardingTime","avgBoardingTime","avgBoardingTime",
              "seatShuffle","seatShuffle","seatShuffle","seatShuffle","seatShuffle","seatShuffle","seatShuffle","seatShuffle",
              "numStops","numStops","numStops","numStops","numStops","numStops","numStops","numStops",
              "cumWaitTime","cumWaitTime","cumWaitTime","cumWaitTime","cumWaitTime","cumWaitTime","cumWaitTime","cumWaitTime"]
```

```python
order_models = ["btf","ftb","random","sm","sp","wta_btf","wta_ftb","wta_random","btf","ftb","random","sm","sp","wta_btf","wta_ftb","wta_random","btf","ftb","random","sm","sp","wta_btf","wta_ftb","wta_random","btf","ftb","random","sm","sp","wta_btf","wta_ftb","wta_random","btf","ftb","random","sm","sp","wta_btf","wta_ftb","wta_random"]
for i in range(len(order_models)):
        order_models[i] = stats_list[i]+"_"+order_models[i]
#order_models
```

```python
excel_rename = []
for i in excel_reorder:
    i.columns=i.columns[:0].tolist() + order_models
    excel_rename.append(i)
#excel_rename[0]
```
